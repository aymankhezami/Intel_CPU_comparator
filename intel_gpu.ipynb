```python
extract = "HDMI:|DP:|eDP:|Hz"
df["max_resolution"]=df["max_resolution"].str.replace(extract,"",regex=True)


delimiters = r"@|\n"
split_columns = ["max_resolution_hdmi", "freq_hdmi(en Hz)", "max_resolution_dp", "freq_dp(en Hz)", "max_resolution_edp", "freq_edp"]
df[split_columns]= df["max_resolution"].str.split(delimiters, expand=True,n=5)

df.drop("max_resolution",axis=1,inplace=True)
```

### Modifier processor_number

```python
df["nGeneration"] = df["nGeneration"].str.replace(" ","")
```

```python
df["X-series"]=df["nGeneration"].apply(lambda x: "Yes"if x=="X-series" else "No")
df["X-series"].unique()
```

```python
df["nGeneration2"] = df["processor_number"].str.extract(r'-(\d{2})')
df["nGeneration2"].unique()
```

```python
def process_number(num):
    if pd.notna(num):  # Check if the value is not NaN
        num_str = str(int(num))
        if num_str in ['13', '12', '11', '10']:
            return num_str
        else:
            return num_str[0] + num_str[2:]
    else:
        return np.NaN

df["nGeneration2"] = df["nGeneration2"].apply(process_number)
```

```python
df["processor_number"] = df["processor_number"].str.replace('+', '-')
```

```python
df["nGeneration2"] = pd.to_numeric(df["nGeneration2"], errors='coerce')  
```

```python
df["series"] = df["processor_number"].str.extract(r'.*0(.*)')
```

```python
df["series"].unique()
```

```python
dict_repalce_series = {'k':'K','T\xa0':'T','KF\xa0':'KF','K\xa0':'K','325':'O',
                       '5T':'T','5':'O','5F':'F','':'O','86K':'K','1TE':'TE',
                       '1E':'E','2P':'P','98P':'P','XE':'E','X':'O',}
df["series"].replace(dict_repalce_series,inplace=True)
df["series"].unique()
```
